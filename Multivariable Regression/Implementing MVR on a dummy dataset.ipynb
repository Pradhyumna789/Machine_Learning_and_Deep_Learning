```python
from sklearn import datasets
boston = datasets.load_boston()
X = boston.data
Y = boston.target
print(X.shape)
```

```python
import pandas as pd
X = pd.DataFrame(X)
X.columns = boston.feature_names
X.describe()
```

```python
X2 = X.copy()
```

```python
for feature in X:
    new_feature = feature + "_" + feature
    X2[new_feature] = X[feature]**2
    
X2.describe()
```

```python
X3 = X.copy()
```

```python
for feature1 in X:
    for feature2 in X:
        new_feature = feature1 + '_' + feature2
        X3[new_feature] = X[feature1]*X[feature2]

X3.describe()
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size = 0.3, random_state = 42)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2, Y, test_size = 0.3, random_state = 42)
X3_train, X3_test, Y3_train, Y3_test = model_selection.train_test_split(X3, Y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
algo1 = LinearRegression(normalize = True)
algo2 = LinearRegression(normalize = True)
algo3 = LinearRegression(normalize = True)
```

```python
algo1.fit(X_train, Y_train)
algo2.fit(X2_train, Y2_train)
algo3.fit(X3_train, Y3_train)
```

```python
X_train_score = algo1.score(X_train, Y_train)
X_test_score = algo1.score(X_test, Y_test)
X2_train_score = algo2.score(X2_train, Y2_train)
X2_test_score = algo2.score(X2_test, Y2_test)
X3_train_score = algo3.score(X3_train, Y3_train)
X3_test_score = algo3.score(X3_test, Y3_test)
```

```python
print(X_test_score)
print(X2_test_score)
print(X3_test_score)
```
