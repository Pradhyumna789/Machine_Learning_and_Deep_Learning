# Customizing Graph

```python
import numpy as np
x = np.arange(0, 5, 0.1)
y = x ** 3
plt.plot(x, y, color = "purple", marker = "o")
plt.show()
```

```python
import numpy as np

x = np.arange(0, 5, 0.1)
y = x ** 3
y2 = x ** 2

plt.plot(x, y, color = "black", linewidth = 5, label = "x^3") # call legend() functions after defining label
plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")


plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.title("Matplotlib demo")

plt.show()
```

```python
# Changing axes
import numpy as np

x = np.arange(0, 5, 0.1)
y = x ** 3
y2 = x ** 2

plt.plot(x, y, color = "black", linewidth = 5, label = "x^3") # call legend() functions after defining label
plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

plt.axis([0, 10, 0, 200]) # 0 to 100 for 'x' axis and 0 to 10 for 'y' axis

plt.grid() # created a grid
plt.text(2, 80, "Text", fontsize = 12) # Writes 'Text' at (x = 2, y = 80)

plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.title("Matplotlib demo")

plt.show()
```

# Bubble Chart

```python
import matplotlib.pyplot as plt

year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
colors = ['red', 'blue', 'violet', 'yellow', 'green', 'black']

plt.scatter(year, salary, s = population, c = colors)
plt.title('Bubble Chart')
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = 'r', alpha = 0.3) # decrease in alpha = more transparency
plt.title('Bubble Chart')
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
c[1] = c[0]

plt.scatter(year, salary, s = population, c = c, alpha = 0.3, edgecolor = 'red', marker = '^') # decrease in alpha = more transparency
plt.title('Bubble Chart')
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
c[1] = c[0]

plt.scatter(year, salary, s = population, c = c, alpha = 0.3, edgecolor = 'red', marker = '^') # decrease in alpha = more transparency
plt.title('Bubble Chart')
plt.text(year[0] + 0.1, salary[0] + 0.1, population[0])
plt.text(year[4] + 0.1, salary[4] + 0.1, population[4])
plt.show()
```

# Pie Chart

```python
# pie graphs
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ['blue', 'purple', 'red', 'pink']
explode = [0.1, 0, 0.2, 0]
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct = '%.1f', startangle = 100) # or autopct = "%d"
plt.title("Split among classes")
plt.axis('equal')
plt.show()
```

```python
# pie graphs
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ['blue', 'purple', 'red', 'pink']
explode = [0.1, 0, 0.2, 0]
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct = '%.1f', counterclock = False, startangle = 100) # or autopct = "%d"
plt.title("Split among classes")
plt.axis('equal')
plt.show()
```

# Histogram

- Amazing way to look at the distribution of data
- Best way to look at the frequency of data(like how many times a particular range of data is occuring)

```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,1,3,4,2,1,1]
```

```python
plt.hist(a)
plt.show()
```

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
```

```python
plt.hist(b, edgecolor = 'black')
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
# Now we want to see which number's frequencies are clubbed together
import numpy as np

xt = np.arange(22)
plt.hist(b, edgecolor = 'black')
plt.xticks(xt)
plt.show()

# 1 to 3.1 - first bin
# 3.1 to 5.2 - second bin
# Likewise it creates 20 bins - (max - min) = (21 - 1) = 20 bins in total
# Size of each bin - [(max - min) / (total bins)] = [(22 - 1) / (10)] = 2.1 is the width of each bin
```

```python
# We can decide how many bins do we want to have
# Width of each bin = [(22 - 1) / (21)] = 1
import numpy as np

xt = np.arange(22)
plt.hist(b, bins = 21, edgecolor = 'black')
plt.xticks(xt)
plt.show()
# [1,2), [2, 3), ..., [20, 21), [21, 22]
```

```python
plt.hist(b, bins = [1,3,6,10,15,20])
plt.show()
# [1,3)
# [3,6)
# [6,10)
# [10,15)
# [15,20)
```

```python
xt = np.arange(21)
plt.xticks(xt)
plt.hist(b, bins = [1,3,6,10,15,20])
plt.axis([0,25,0,10]) #[xmin, xmax, ymin, ymax]
plt.show()
```
